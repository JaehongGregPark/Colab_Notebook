이 문제에서는 가위바위보 게임의 업그레이드 버전을 함수로 작성해야 합니다. 이번에는 몇 판을 진행할 것인지 입력을 받아야 하고, 사용자로부터 "가위", "바위", "보" 또는 0, 1, 2 중에서 선택하도록 해야합니다. 이외에 다른 입력이 들어오면 재입력을 요청해야 합니다. 게임 종료 후에는 나와 컴퓨터의 총 전적을 출력해야 합니다.

함수를 정의할 때는 'def' 키워드를 사용하고, 함수 이름과 매개변수를 지정합니다. 함수 내에서는 조건문과 반복문을 사용하여 게임을 진행하고 전적을 저장하고 출력합니다.
구현 순서는 다음과 같습니다 :
1. 가위바위보 게임 함수 정의하기
2. 게임 진행 횟수 입력받기
3. 전적을 저장할 변수 초기화하기
4. 가위바위보 선태 옵션 정의하기
5. 입력된 횟수만큼 게임 진행하기
6. 사용자의 선택 입력받기
7. 잘못된 입력이 들어올 경우 재입력 요청하기
8. 사용자의 선택을 숫자에서 실제 가위바위보로 변환하기
9. 컴퓨터의 선택을 랜덤으로 생성하기
10. 사용자와 컴퓨터의 선택 출력하기
11. 승패 결과 출력하기
12. 전체 전적 출력하기
13. 가위바위보 게임 함수 실행하기

```python
import random

def rps_game():
    # 게임 진행 횟수 입력받기
    num_games = int(input("몇 판을 진행하시겠습니까? "))

    # 전적을 저장할 변수 초기화하기
    user_wins = 0
    computer_wins = 0

    # 가위바위보 선태 옵션 정의하기
    options = {0: "가위", 1: "바위", 2: "보"}

    for game in range(num_games):
        print(f"\n{game+1}번째 판을 시작합니다.")

        # 입력 유효성 검사를 위한 반복문
        while True:
            # 사용자의 선택 입력받기
            user_choice = input("가위(0), 바위(1), 보(2) 중에서 선택하세요: ")

            # 잘못된 입력이면 재입력 요청
            if user_choice not in ["0", "1", "2"]:
                print("잘못된 입력입니다. 0, 1, 2 중에서 선택해주세요.")
                continue
            else:
                break

        # 사용자의 선택을 숫자에서 실제 가위바위보로 변환하기
        user_choice = options[int(user_choice)]

        # 컴퓨터의 선택을 랜덤으로 생성하기
        computer_choice = options[random.randint(0, 2)]

        # 사용자와 컴퓨터의 선택 출력하기
        print(f"사용자: {user_choice}, 컴퓨터: {computer_choice}")

        # 승패 결과 출력하기
        if user_choice == computer_choice:
            print("비겼습니다!")
        elif (
            (user_choice == "가위" and computer_choice == "보") or
            (user_choice == "바위" and computer_choice == "가위") or
            (user_choice == "보" and computer_choice == "바위")
        ):
            print("사용자가 이겼습니다!")
            user_wins += 1
        else:
            print("컴퓨터가 이겼습니다!")
            computer_wins += 1

    # 전체 전적 출력하기
    print(f"\n--- 총 전적 ---")
    print(f"사용자: {user_wins}승, 컴퓨터: {computer_wins}승")

# 가위바위보 게임 함수 실행하기
rps_game()
```
