구현 순서는 다음과 같습니다 :
1. 가위바위보 게임 함수 정의하기
2. 게임 진행 횟수 입력받기

4. 가위바위보 선택 옵션 정의하기
5. 입력된 횟수만큼 게임 진행하기
6. 사용자의 선택 입력받기
7. 잘못된 입력이 들어올 경우 재입력 요청하기

9. 컴퓨터의 선택을 랜덤으로 생성하기
10. 사용자와 컴퓨터의 선택 출력하기
11. 승패 결과 출력하기
12. 전체 전적 출력하기
13. 가위바위보 게임 함수 실행하기

```python
import random

def rps():
    # 게임 진행 횟수 입력받기
    num_games = int(input("몇 판을 진행하시겠습니까? "))

    # 전적을 저장할 변수 초기화하기
    user_wins = 0
    computer_wins = 0

    # 가위바위보 선태 옵션 정의하기
    options = {0: "가위", 1: "바위", 2: "보"}

    for game in range(num_games):

        print(f"\n{game+1}번째 판을 시작합니다.")

        # 입력 유효성 검사를 위한 반복문
        while True:
            # 사용자의 선택 입력받기
            user_choice = input("가위(0), 바위(1), 보(2) 중에서 선택하세요: ")

            # 잘못된 입력이면 재입력 요청
            if user_choice not in ["0", "1", "2"]:
                print("잘못된 입력입니다. 0, 1, 2 중에서 선택해주세요.")
                continue
            else:
                break

        # 사용자의 선택을 숫자에서 실제 가위바위보로 변환하기
        user_choice = options[int(user_choice)]

        # 컴퓨터의 선택을 랜덤으로 생성하기
        computer_choice = options[random.randint(0, 2)]

        # 사용자와 컴퓨터의 선택 출력하기
        print(f"사용자: {user_choice}, 컴퓨터: {computer_choice}")

        # 승패 결과 출력하기
        if user_choice == computer_choice:
            print("비겼습니다!")
        elif (
            (user_choice == "가위" and computer_choice == "보") or
            (user_choice == "바위" and computer_choice == "가위") or
            (user_choice == "보" and computer_choice == "바위")
        ):
            print("사용자가 이겼습니다!")
            user_wins += 1
        else:
            print("컴퓨터가 이겼습니다!")
            computer_wins += 1

    # 전체 전적 출력하기1
    131

    print(f"\n--- 총 전적 ---")
    print(f"사용자: {user_wins}승, 컴퓨터: {computer_wins}승")

# 가위바위보 게임 함수 실행하기
rps()
```

           그리고 중앙값도 함께 출력 해봅시다.
           (단, 중앙값이 짝수가 아닐 경우에는 중앙값은 출력을 하지 않고 짝수인 수만 출력한다)
           [난이도 : ⭐️⭐️/5]
         

중앙값 : 정중앙에 있는 값

특정 두 숫자 사이의 수들을 리스트로 만드는 법

이 문제에서는 입력으로 받은 두 숫자 사이에 있는 짝수를 출력하는 함수를 작성해야 합니다. 또한 그 중앙값도 함께 출력해야하죠. 중앙값은 입력된 두 숫자 사이에 가운데 값으로 정의됩니다. 그러나 중앙값이 짝수인 경우에만 출력하며, 홀수인 경우에는 출력하지 않습니다.

함수를 정의할 때는 'def' 키워드를 사용하고, 함수 이름과 매개변수를 지정합니다. 함수 내에서는 주어진 두 숫자 사이의 수들을 리스트로 만들고, 그 중에서 짝수를 필터링하여 출력합니다. 또한, 중앙값이 짝수인 경우에만 출력합니다.

구현 순서는 다음과 같습니다 :
1. 함수 정의하기
2. 주어진 두 숫자 사이의 수들을 리스트로 생성하기
3. 중앙값 계산하기
4. 중앙값이 짝수인 경우에만 출력하기
5. 짝수인 숫자들을 필터링하여 출력하기
6. 사용자로부터 두 숫자를 입력받아 함수 실행하기

프로그램 실행의 결과는 아래와 같이 출력되어야 합니다.

```python
def even_numbers(num1,num2):
  # 주어진 두 숫자 사이의 수들을 리스트로 생성하기
    numbers_between = list(range(min(num1, num2) + 1, max(num1, num2)))

    # 중앙값 계산하기
    middle_index = len(numbers_between) // 2
    median = numbers_between[middle_index] if len(numbers_between) % 2 != 0 else None

    # 중앙값이 짝수인 경우에만 출력하기
    if median is not None and median % 2 == 0:
        print("중앙값:", median)

    # 짝수인 숫자들을 필터링하여 출력하기
    even_numbers = [num for num in numbers_between if num % 2 == 0]
    print("짝수:", even_numbers)

# 사용자로부터 두 숫자를 입력받아 함수 실행하기
try:
  i1 = int(input("첫번째수입력:"))
  i2 = int(input("두번째수입력:"))
  even_numbers(i1,i2)

except ValueError:
    print("잘못된 입력입니다. 숫자를 입력해주세요.")
```

Q4. 2개의 숫자 사이에 소수가 몇 개인지 출력하는 함수를 만들어 봅시다. [난이도 : ⭐️⭐️⭐️/5]

소수: 1과 자기 자신만을 약수로 가지는 수

이 문제에서는 입력으로 받은 두 숫자 사이에 있는 소수의 개수를 출력하는 함수를 작성해야 합니다. 소수는 '1과 자기 자신만을 약수로 가지는 수'이죠.

함수를 정의할 때는 'def' 키워드를 사용하고, 함수 이름과 매개변수를 지정합니다. 함수 내에서는 주어진 두 숫자 사이에 있는 모든 수들을 순회하면서 그 숫자가 소수인지 확인합니다. 소수인 경우, 카운트를 증가시키며, 마지막에는 카운트된 소수의 개수를 출력합니다.

구현 순서는 다음과 같습니다 :
1. 함수 정의하기
2. 소수 개수를 카운트할 변수 초기화하기
3. 주어진 두 숫자 사이의 수들을 순회하면서 소수인지 확인하기
4. 소수 확인을 위한 반복문 작성하기
5. 소수인 경우 카운트 증가시키기
6. 최종 소수 개수 출력하기
7. 사용자로부터 두 숫자를 입력받아 함수 실행하기


```python
def prime(i1,i2):
  prime_cnt = 0
  prime_yn = True
  for i in range(i1, i2+1):
    prime_yn = False
    for j in range(2,i):
      if i%j == 0 :
        prime_yn = True

      if prime_yn:
        prime_cnt = prime_cnt+1
        print("{}는 소수입니다".format(i))
      else:
        print("{}는 소수가 아닙니다".format(i))

  print("소수의 개수는 {}".format(prime_cnt))

#i1 = int(input("첫번째수입력:"))
#i2 = int(input("두번째수입력:"))
#prime(i1,i2)

def is_prime(num):
    # 소수 확인을 위한 반복문 작성하기
    if num < 2:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    #print(num,"은 소수입니다")
    return True


def count_primes(start, end):
    # 소수 개수를 카운트할 변수 초기화하기
    count = 0

    # 주어진 두 숫자 사이의 수들을 순회하면서 소수인지 확인하기
    for num in range(start, end + 1):
        # 소수인 경우, 카운트 증가시키기
        if is_prime(num):
            count += 1

    # 최종 소수 개수 출력하기
    return count


# 사용자로부터 두 숫자를 입력받아 함수 실행하기
try:
    i1 = int(input("첫 번째 숫자를 입력하세요: "))
    i2 = int(input("두 번째 숫자를 입력하세요: "))

    if i1 >= i2:
        print("두 숫자는 순서대로 입력되어야 합니다.")
    else:
        result = count_primes(i1, i2)
        print("소수 개수:" ,result)

except ValueError:
    print("잘못된 입력입니다. 숫자를 입력해주세요.")
```
