```python
def grader(i_name,i_score):
  i_score = int(i_score)
  print("학생이름:"+i_name)
  print("점수:"+str(i_score))

  if (i_score >= 95) :
    grade = 'A+'
  elif(i_score >= 90) :
    grade = 'A'
  elif(i_score >= 85) :
    grade = 'B+'
  elif(i_score >= 80) :
    grade = 'B'
  elif(i_score >= 75) :
    grade = 'C+'
  elif(i_score >= 70) :
    grade = 'C'
  elif(i_score >= 65) :
    grade = 'D+'
  elif(i_score >= 60) :
    grade = 'D'
  elif(i_score >= 0) :
    grade = 'F'

  print("학점:"+grade)


i_name, i_score = input("input").split(",")

grader(i_name,i_score)
```
