4. 면담 대상자의 평균 실적이 3보다 크면, 면담 대상자에서 제외됩니다.



# 이름, 실적
member_names = ["갑돌이", "갑순이", "을돌이", "을순이", "병돌이", "병순이"]
member_records = [[4,5,3,5,6,5,3,4,1,3,4,5],[2,3,4,3,1,2,0,3,2,5,7,2],
           [1,3,0,3,3,4,5,6,7,2,2,1],[3,2,9,2,3,5,6,6,4,6,9,9],

```python
member_names = ["갑돌이", "갑순이", "을돌이", "을순이", "병돌이", "병순이"]
member_records = [
    [4, 5, 3, 5, 6, 5, 3, 4, 1, 3, 4, 5],
    [2, 3, 4, 3, 1, 2, 0, 3, 2, 5, 7, 2],
    [1, 3, 0, 3, 3, 4, 5, 6, 7, 2, 2, 1],
    [3, 2, 9, 2, 3, 5, 6, 6, 4, 6, 9, 9],
    [8, 7, 7, 5, 6, 7, 5, 8, 8, 6, 10, 9],
    [7, 8, 4, 9, 5, 10, 3, 3, 2, 2, 1, 3]
]

bonus_candidates = []  # 보너스 대상자 후보 리스트
interview_candidates = []  # 면담 대상자 후보 리스트

# 보너스 대상자와 면담 대상자 구분 및 조건 확인
for name, records in zip(member_names, member_records):
    average_score = sum(records) / len(records)

    if average_score > 5:
        if average_score <= 5:  # 평균 실적이 5보다 크지 않으면 보너스 대상자에서 제외
            continue
        bonus_candidates.append(name)
    elif average_score > 3:
        interview_candidates.append(name)

# 결과 출력
print("보너스 대상자 후보:", bonus_candidates)
print("면담 대상자 후보:", interview_candidates)
```

Q3. 최근 철수는 예금 금리가 낮은 현재 상황을 보고, 주식 시장에 흥미를 느끼게 되었습니다. 그래서 아래와 같이 매수한 종목들의 정보를 활용해 수익률을 계산하고 싶어졌습니다. 철수는 매수한 종목의 이름, 수량, 매수 평균 금액을 알고 있으며, 각 종목의 판매가도 확인했습니다. 이제 우리는 철수를 도와서 종목별 수익률을 계산하는 프로그램을 만들어보려고 합니다. 파이썬을 이용해 종목별 수익률을 계산하고, 수익률이 높은 순서대로 출력해보세요. (소수 둘째자리까지 출력) [난이도 : ⭐️⭐️⭐️/5]


stocks = "삼성전자/10/85000,카카오/15/130000,LG화학/3/820000,NAVER/5/420000"
sells = [82000, 160000, 835000, 410000]

```python
stocks = "삼성전자/10/85000,카카오/15/130000,LG화학/3/820000,NAVER/5/420000"
sells = [82000, 160000, 835000, 410000]

# 종목 정보 파싱하여 딕셔너리 리스트 생성
stock_info = []
for stock in stocks.split(','):
    name, quantity, avg_price = stock.split('/')
    stock_info.append({'name': name, 'quantity': int(quantity), 'avg_price': int(avg_price)})

# 종목별 수익률 계산 및 저장
for i, sell_price in enumerate(sells):
    stock_info[i]['sell_price'] = sell_price
    investment = stock_info[i]['quantity'] * stock_info[i]['avg_price']
    earnings = (stock_info[i]['sell_price'] * stock_info[i]['quantity']) - investment
    stock_info[i]['earnings'] = earnings
    stock_info[i]['roi'] = (earnings / investment) * 100  # 수익률 계산

# 수익률이 높은 순서대로 정렬
sorted_stock_info = sorted(stock_info, key=lambda x: x['roi'], reverse=True)

# 결과 출력
print("종목별 수익률 (소수 둘째자리까지):")
for stock in sorted_stock_info:
    print(f"{stock['name']} - 수익률: {stock['roi']:.2f}%")
```
