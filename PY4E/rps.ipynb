```python
import random
computer = random.randint(0, 2)

if computer == 0 :
  computer = '가위'
elif computer == 1 :
  computer = '바위'
elif computer == 2 :
  computer = '보'


my = input("가위바위보")

print("나:"+my)

print("computer:"+computer)

if computer == '가위':
    if my == '가위':
        print('무승부')
    elif my == '바위':
        print('플레이어 승리')
    elif my == '보':
        print('컴퓨터 승리')
elif computer == '바위':
    if my == '가위':
        print('컴퓨터 승리')
    elif my == '바위':
        print('무승부')
    elif my == '보':
        print('플레이어 승리')
elif computer == '보':
    if my == '가위':
        print('플레이어 승리')
    elif my == '바위':
        print('컴퓨터 승리')
    elif my == '보':
        print('무승부')
```
