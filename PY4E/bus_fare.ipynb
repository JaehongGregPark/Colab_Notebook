```python
def bus_fare(i_age,i_type):
  age = int(i_age)
  bus_fee = 0

  if (age<8) :
    bus_fee = 0
    print("a")
  elif (age<14) :
    print("b")
    if (i_type=="현금"):
      bus_fee = 450
      print("b1")
    elif(i_type=="카드"):
      bus_fee = 450
      print("b2")
  elif (age<20) :
    print("c")
    if (i_type=="현금"):
      bus_fee = 1000
      print("c1")
    elif(i_type=="카드"):
      bus_fee = 720
      print("c2")
  elif (age<75) :
    print("d")
    if (i_type=="현금"):
      bus_fee = 1300
      print("d1")
    elif(i_type=="카드"):
      bus_fee = 1200
      print("d2")
  elif (age>=75) :
      bus_fee = 0
      print("e")

  print("age="+str(age))
  print("bus_fee="+str(bus_fee))

  if bus_fee == 0 :
    str_bus_fee = "무료"
  else :
    str_bus_fee = str(bus_fee) + "원"

  print("나이:" + str(age) + '세')
  print("지불유형:"+ i_type)
  print("버스요금:"+str_bus_fee)

i_age, i_type = input("input").split(",")
bus_fare(i_age,i_type)
```
