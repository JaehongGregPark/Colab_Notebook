```python
def yearly_payment(monthly_payment):
  year_salary = monthly_payment*12
  print("세전연봉: "+ str(year_salary) +"만원" )
  if(year_salary<1200):
    year_pay = year_salary - (year_salary * 0.06)
  elif(year_salary<4600):
    year_pay = year_salary - (year_salary * 0.15)
  elif(year_salary<8800):
    year_pay = year_salary - (year_salary * 0.24)
  elif(year_salary<15000):
    year_pay = year_salary - (year_salary * 0.35)
  elif(year_salary<30000):
    year_pay = year_salary - (year_salary * 0.38)
  elif(year_salary<50000):
    year_pay = year_salary - (year_salary * 0.4)
  elif(year_salary>50000):
    year_pay = year_salary - (year_salary * 0.42)
  year_pay = int(year_pay)
  print("세후연봉: "+ str(year_pay) +"만원" )

monthly_payment = int(input("monthly_payment="))
yearly_payment(monthly_payment)
```
