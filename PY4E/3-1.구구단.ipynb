이 문제에서는 사용자로부터 숫자를 입력 받은 뒤, 해당 숫자에 대한 구구단을 출력하는 함수를 작성해야 합니다.
출력은 조건 1과 조건 2를 모두 만족해야 합니다.

 #조건 1: 홀 수 번째만 출력하기
 #조건 2: 값이 50이하인 것만 출력하기

프로그램을 작성할 때, 함수를 정의하고 사용자로부터 숫자를 입력받는 방법, 구구단을 출력하는 방법, 그리고 조건 1과 조건 2를 만족시키는 로직을 작성해야 합니다.

함수를 정의할 때는 'def' 키워드를 사용하고, 함수 이름을 지정한 뒤 함수의 매개변수로 숫자를 받습니다. 구구단을 출력하는 부분에서는 반복문을 사용하여 숫자 1부터 9까지를 곱한 값을 출력합니다. 이때 조건문을 추가하여 홀 수 번째의 값만 출력하고, 값이 50이하인 것만 출력하도록 설정합니다.




```python
def gugudan(a):
  print(a,"단")
  for x in range(a, a+1):
    for y in range(1,9+1):

 #조건 1: 홀 수 번째만 출력하기
 #조건 2: 값이 50이하인 것만 출력하기
      if(y%2==1) and (x*y <= 50):
        print(x, '*', y, '=', x * y )

a = int(input("구구단수입력:"))
gugudan(a)
```
