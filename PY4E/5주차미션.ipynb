Q1. 여러분 혹시 베스킨라빈스31 게임을 아시나요? 1부터 31까지 숫자를 플레이어들끼리 번갈아 외치다가 31을 외치는 사람이 패배하는 게임인데요. 파이썬 함수를 활용해 다음의 규칙을 만족하는 게임을 만들어보고, 컴퓨터와 대결을 해보세요. [난이도 : ⭐️⭐️⭐️/5]

<게임 규칙>

1. 플레이어는 매 턴마다 숫자를 입력하며, 입력된 숫자들은 공백(space)로 구분합니다.
2. 플레이어와 컴퓨터 모두 한 번의 턴에서 최대 3개의 숫자를 외칠 수 있습니다.
3. 숫자를 외칠 때에는 이전에 외쳐진 숫자보다 1 큰 수만 외칠 수 있습니다.
   - 예를 들어, 컴퓨터가 마지막 숫자로 7을 외쳤다면, 플레이어는 7을 이어서 8, 9, 10 까지 외칠 수 있습니다.
   - 컴퓨터도 동일한 규칙을 따라야 합니다.
4. '31'이라는 숫자가 나오면 게임을 종료합니다.

✔︎ 보조 설명 :

이번 문제는 베스킨 라빈스 31이라는 숫자 게임을 파이썬 프로그래밍으로 풀어보는 문제입니다.
숫자의 크기를 1씩 늘려가며 이어가는 게임으로, 누가 먼저 31을 외칠 것인지 예상하며
외칠 숫자의 개수를 조정하는 재미가 있는 게임입니다.

이 문제는 바로 앞선 시도의 마지막 숫자를 기억해야는 점이 특징입니다.
그리고 컴퓨터가 숫자를 외칠 때, 1개가 아닌 최대 3개까지 외칠 수 있다는 점을 주요히 봐야합니다.

구현을 위한 작성 순서는 다음과 같습니다. 이외에도 다양한 방법이 있으므로 자유롭게 코드를 작성해보세요!

1. 플레이어와 컴퓨터가 숫자를 외치는 함수를 정의하고, 게임 진행을 위한 초기 변수를 설정합니다.
2. 반복문을 사용하여 게임이 종료될 때까지 게임을 반복해서 실행합니다.
3. 플레이어의 턴인지, 컴퓨터의 턴인지 판단하여 숫자를 입력받거나 출력합니다.
4. 현재 숫자가 31이 되었을 때, 외친 사람이 패배한 것으로 게임을 종료합니다.
5. 최종 결과를 출력합니다.

```python
import random

def player_turn(current_number):
    """
    플레이어의 턴을 처리하는 함수입니다.
    :param current_number: 현재 숫자
    :return: 업데이트된 현재 숫자와 플레이어가 선택한 숫자들의 리스트
    """
    while True:
        try:
            nums = input("My turn - 숫자를 입력하세요: (공백으로 구분): ").split()
            nums = [int(num) for num in nums]

            valid_input = True
            for num in nums:
                if num != current_number + 1:
                    valid_input = False
                    break
                current_number = num

            if valid_input:
                return current_number, nums
            else:
                print("유효하지 않은 입력입니다. 1씩 증가해야 합니다.")
        except ValueError:
            print("유효하지 않은 입력입니다. 정수를 입력하세요.")

def computer_turn(current_number):
    """
    컴퓨터의 턴을 처리하는 함수입니다.
    :param current_number: 현재 숫자
    :return: 업데이트된 현재 숫자와 컴퓨터가 선택한 숫자들의 리스트
    """
    # 남은 숫자 범위를 계산합니다.
    # 현재 숫자에서 31을 뺀 값은 게임에서 목표로 하는 숫자인 31까지 남은 숫자의 개수를 나타냅니다.
    # 컴퓨터가 선택할 수 있는 숫자의 최대 개수를 3으로 제한합니다. 만약 남은 숫자 범위가 3 이하라면,
    # 컴퓨터는 그 범위 안에서 숫자를 선택하게 됩니다.
    max_count = min(3, 31 - current_number)

    #. 이 값은 컴퓨터가 선택할 숫자의 개수를 의미합니다.
    # 컴퓨터는 count 개수만큼 연속된 숫자를 선택하여 게임을 진행합니다.
    # 이렇게 하면 컴퓨터가 1부터 max_count 범위 내에서 랜덤하게 숫자를 선택 할 수 있습니다.
    count = random.randint(1, max_count)

    #count 개수만큼 숫자들을 생성하여 리스트 nums에 저장합니다.
    # current_number는 현재 숫자를 나타내며, i는 숫자 생성의 반복 횟수입니다.
    # 따라서 current_number + i + 1은 연속된 숫자를 생성하여 리스트 nums에 저장하는 것을 의미합니다.
    nums = [current_number + i + 1 for i in range(count)]

    current_number += count
    return current_number, nums

def bs31():
    """
    게임을 진행하는 메인 함수입니다.
    """
    current_number = 0
    while current_number < 31:
        print(f"현재 숫자: {current_number}")

        # 플레이어의 턴
        current_number, nums = player_turn(current_number)
        print(f"My turn: {nums}")

        if current_number >= 31:
            break

        # 컴퓨터의 턴
        current_number, nums = computer_turn(current_number)
        print(f"Computer turn: {nums}")

        if current_number >= 31:
            break

    # 승자 결정
    winner = "플레이어" if current_number == 31 else "컴퓨터"
    print(f"{winner} 승리! 최종 숫자는 {current_number}입니다.")

# 게임 실행
bs31()
```

✔︎ 보조 설명 :

이번 중간고사는 총 10문제로 구성되어 있습니다. 학생들은 열심히 공부를 하고 문제에 대한 답안을 작성했습니다.
그리고 선생님도 정답지를 만들어 각 문제에 대한 정답을 기록했습니다.
이제 학생들의 답안과 정답지를 비교하여 점수를 계산하고, 각 학생의 점수와 등수를 출력해야 합니다.

학생들의 답안은 's' 리스트에 이름과 답안 문자열로 주어졌습니다.

정답은 'a' 리스트로 주어지며, 각 문제에 대한 정답은 숫자로 나타내어져 있습니다.

다음은 구현을 위한 작성 순서입니다. 이외에도 다양한 방법이 있으니 자유롭게 코드를 작성해보세요!

1. 학생들의 답안과 정답을 각 변수에 문자열의 형태로 저장합니다.
2. 채점 함수를 만듭니다.
   - 채점 함수는 학생의 답안과 정답을 비교하여 점수를 계산합니다.
   - 각 문제에 대해 학생의 답안과 정답을 비교해, 일치하면 해당 문제의 점수를 학생의 총점에 더합니다.
   - 모든 문제를 채점한 후, 학생별 총 점수를 반환합니다.
3. 모든 학생들의 점수 계산을 완료한 후, 등수를 매깁니다.
   - 학생들의 점수와 이름을 쌍으로 묶은 리스트를 생성합니다.
   - 리스트를 점수 기준으로 내림차순으로 정렬합니다.
4. 최종적으로 학생들의 점수와 등수를 출력합니다.

```python
def calculate_score(answer, key):
    score = 0
    for i in range(len(key)):
        if answer[i] == str(key[i]):
            score += 10
    return score

def grader(s, a):
    student_scores = []

    # 학생별로 답안 채점 및 점수 계산
    for student in s:
        name, answer = student.split(',')
        score = calculate_score(answer, a)
        student_scores.append((name, score))

    # 점수를 기준으로 내림차순으로 학생들 정렬
    def sort_by_score(student):
        return student[1]
    #학생들의 점수를 기준으로 내림차순으로 정렬된 학생들의 이름과 점수가 저장.
    sorted_students = sorted(student_scores, key=sort_by_score, reverse=True)

    # 등수 계산 및 출력
    for rank, (name, score) in enumerate(sorted_students, start=1):
        if rank == 1:
            rank_str = "1등"
        else:
            rank_str = f"{rank}등"
        print(f"학생: {name} 점수: {score}점 {rank_str}")

# 학생 답안과 정답 리스트
s = ["김갑,3242524215",
     "이을,3242524223",
     "박병,2242554131",
     "최정,4245242315",
     "정무,3242524315"]

a = [3, 2, 4, 2, 5, 2, 4, 3, 1, 2]

grader(s, a)
```

<게임 규칙>

1. 게임에서 사용되는 숫자는 숫자는 0부터 100까지의 정수로 한정합니다.

   - 만약 사용자의 숫자가 정답보다 작다면, "Up!"이라는 메세지를 출력합니다.

    - 만약 앞선 시도에서 작성한 숫자보다 크거나 작은 경우, 올바른 범위로 예측할 수 있게 가이드합니다.
4. 게임이 종료되면, 정답과 함께 사용자가 정답을 맞추기까지 시도한 횟수를 출력합니다.

 ✔︎ 보조 설명 :

이번 문제는 익히 알고 계시는 Up&Down 문제를 파이썬 프로그래밍으로 풀어보는 문제입니다.
컴퓨터가 선택한 랜덤 숫자를 맞추는 게임으로, 예측 범위를 점점 좁혀가는 재미가 있는 게임입니다.

이 문제는 사용자가 입력한 모든 예측 숫자를 기억해야는 점이 특징입니다.
앞선 시도들을 리스트에 저장하고, 다음 시도들과 계속 비교하며 조금 더 수월하게 정답을 맞출 수 있도록 가이드 합니다.

구현을 위한 작성 순서는 다음과 같습니다. 이외에도 다양한 방법이 있으므로 자유롭게 코드를 작성해보세요!

1. 컴퓨터가 0부터 100 사이의 정수 중 하나를 랜덤하게 선택하여 정답을 설정합니다.
2. 사용자로부터 숫자를 입력받는 함수를 자겅합니다.
3. 사용자가 입력한 숫자와 정답을 비교하여 각 상황별로 메세지를 출력하는 함수를 작성합니다.
4. 이전 시도에서 이미 사용한 숫자인지를 확인하는 함수를 작성합니다.
5. 게임을 진행하는 메인 함수를 작성합니다.



```python
import random

def main():
    # 랜덤한 숫자 생성
    number = random.randint(0, 100)
    attempts = []

    print("Up&Down 게임을 시작합니다!\n")

    # 1부터 100까지 시도 횟수를 반복
    for attempt in range(1, 101):
        print(f"{attempt}차 시도")

        # 사용자로부터 숫자를 입력받는 함수 호출
        guess = get_user_input(attempts)
        attempts.append(guess)

        # 입력한 숫자와 랜덤 숫자 비교
        if guess == number:
            print(f"정답입니다. {attempt}차 시도만에 예측에 성공했네요!! 게임을 종료합니다.")
            break
        else:
            # Up 또는 Down 힌트 출력
            if guess < number:
                print("Up! 다시 한번 고민해보세요.")
            else:
                print("Down! 다시 한번 고민해보세요.")

def get_user_input(attempts):
    while True:
        try:
            # 사용자로부터 숫자 입력 받기
            guess = int(input("숫자를 예측해보세요 : "))
            if guess < 0 or guess > 100:
                print("0부터 100 사이의 숫자를 입력하세요.")
            elif is_duplicate_guess(attempts):
                print("이미 사용한 숫자입니다. 다른 숫자를 입력하세요.")
            else:
                return guess
        except ValueError:
            print("유효하지 않은 입력입니다. 정수를 입력하세요.")

def is_duplicate_guess(attempts):
    # 이미 사용한 숫자인지 확인
    return len(attempts) > 0 and attempts[-1] in attempts[:-1]

# 메인 함수 호출
main()
```

✔︎ 보조 설명 :

이번 문제는 파이썬 프로그래밍으로 날짜 계산기를 만들어보는 문제입니다.

이 문제는 '날짜와 요일'을 숫자로 변환해 계산한다는 점이 특징입니다.
입력받은 날짜를 월과 일로 분리하여 정수형으로 저장하고, 계산된 결과가 한달의 일수를 초과하면 월을 증가시켜야 합니다.
같은 방법으로 입력받은 요일을 기준으로 날짜를 계산해야 합니다.

구현을 위한 작성 순서는 다음과 같습니다. 이외에도 다양한 방법이 있으니 자유롭게 코드를 작성해보세요!

1. 사용자로부터 날짜와 요일을 입력받는 함수를 작성합니다.
2. 입력받은 날짜와 요일을 기준으로 100일 후의 날짜와 요일을 계산하는 함수를 작성합니다.
   - 날짜 계산법과 요일 계산법이 다르므로, 두 개를 별개의 함수로 작성합니다.
   - 연도는 고려하지 않으므로, 월이 12를 초과하면 1월로 돌아갑니다.
3. 메인 함수를 작성하여 위 두 함수를 호출하고 결과를 출력합니다.

```python
def get_day_of_week_index(day_of_week):
    """
    입력받은 요일 문자열을 해당 요일의 인덱스로 변환하는 함수입니다.

    Args:
        day_of_week (str): 요일 문자열 (일, 월, 화, 수, 목, 금, 토)

    Returns:
        int: 요일의 인덱스 (0: 일, 1: 월, ..., 6: 토)
    """
    days = ["일", "월", "화", "수", "목", "금", "토"]
    return days.index(day_of_week)

def calculate_d_day(start_month, start_day, start_day_of_week, days_to_add):
    """
    시작 날짜로부터 일정 일 수를 더한 후의 날짜를 계산하는 함수입니다.

    Args:
        start_month (int): 시작 월
        start_day (int): 시작 일
        start_day_of_week (str): 시작 요일 (일, 월, 화, 수, 목, 금, 토)
        days_to_add (int): 더할 일 수

    Returns:
        tuple: 계산된 날짜의 월, 일, 요일의 인덱스 (0: 일, 1: 월, ..., 6: 토)
    """
    # 월과 인덱스를 일치시키기 위해 맨앞 0으로 설정
    days_in_month = [0, 31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

    current_month = start_month
    current_day = start_day
    current_day_of_week_index = get_day_of_week_index(start_day_of_week)

    for _ in range(days_to_add):
        current_day += 1
        current_day_of_week_index = (current_day_of_week_index + 1) % 7

        if current_day > days_in_month[current_month]:
            current_day = 1
            current_month = (current_month + 1) % 13
            if current_month == 0:
                current_month = 1

    return current_month, current_day, current_day_of_week_index

def main():
    # 사용자로부터 시작 날짜와 요일을 입력받습니다.
    start_date = input("시작 날짜(월 일): ")
    start_day_of_week = input("요일: ")

    # 입력받은 시작 날짜를 월과 일로 분리합니다.
    start_month, start_day = map(int, start_date.split())

    # calculate_d_day 함수를 사용하여 100일 뒤의 날짜를 계산합니다.
    d_day_result = calculate_d_day(start_month, start_day, start_day_of_week, 100)

    # 계산된 결과로부터 월, 일, 요일을 추출합니다.
    month, day, day_of_week_index = d_day_result
    day_of_week = ["일", "월", "화", "수", "목", "금", "토"][day_of_week_index]

    # 계산된 결과를 출력합니다.
    print(f"100일 뒤의 날짜는 {month}월 {day}일 {day_of_week}요일입니다.")

main()
```
