마일변환

```python
def miles_to_kilometers(miles):
    kilometers = miles * 1.60934
    return kilometers

try:
    miles_input = float(input("마일을 입력하세요: "))
    kilometers_output = miles_to_kilometers(miles_input)
    print(f"킬로미터: {kilometers_output:.2f} km")
except ValueError:
    print("올바른 숫자를 입력하세요.")
```

인치를 센티미터로

```python
def inches_to_centimeters(inches):
    centimeters = inches * 2.54
    return centimeters

try:
    inches_input = float(input("인치를 입력하세요: "))
    centimeters_output = inches_to_centimeters(inches_input)
    print(f"센티미터: {centimeters_output:.2f} cm")
except ValueError:
    print("올바른 숫자를 입력하세요.")
```

섭씨를 화씨로 변환

```python
def celsius_to_fahrenheit(celsius):
    fahrenheit = (celsius * 9/5) + 32
    return fahrenheit

try:
    celsius_input = float(input("섭씨 온도를 입력하세요: "))
    fahrenheit_output = celsius_to_fahrenheit(celsius_input)
    print(f"화씨 온도: {fahrenheit_output:.2f} °F")
except ValueError:
    print("올바른 숫자를 입력하세요.")
```
