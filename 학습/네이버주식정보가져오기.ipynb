```python
from bs4 import BeautifulSoup
import  requests

def financialData(page) :
  res = requests.get(page)
  soup = BeautifulSoup(res.text,'html.parser')

  body = soup.find('tbody')
  items = body.find_all('tr',{'onmouseover':'mouseOver(this)'})

  for item in items:
    information = item.text
    info = information.split('\n')

    #print(info)
    print(info[1],info[2],info[3],info[6].strip(),info[11].strip(),info[14],info[15],info[16], info[17], info[18])


financeAddress = 'https://finance.naver.com/sise/sise_market_sum.naver?&page='

for i in range(1,44):
  financialData(financeAddress + str(i))
```
