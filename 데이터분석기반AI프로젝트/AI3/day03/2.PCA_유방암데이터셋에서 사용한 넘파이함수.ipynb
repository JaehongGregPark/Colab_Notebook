# np.vstack()


```python
import numpy as np

#두 개의 1차워 배열 생성
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print(a)
print(b)
```

```python
# 두 배열을 수직으로 쌓아서 2차원 배열 생성
result = np.vstack((a,b))
print(result)
```

# np.concatenate()

```python
# 두배열을 가로로 연결해서 1차원 배열 생성
result = np.concatenate((a,b))
print(result)
```

```python
c = np.array([[1, 2], [3, 4]])
d = np.array([[5, 6]])

#두 배열을 세로로 연결해서 2차원 배열 생성
result = np.concatenate((c, d), axis = 0)
print(result)
```

# np.meshgrid()

```python
# x,y축 범위지정
# x축 : -5~5까지 5개의 값
# y축 : -3~3까지 3개의 값

x=np.linspace(-5,5,5)
y=np.linspace(-3,3,3)

print(x)
print(y)

# 격자생성
xx, yy = np.meshgrid(x,y)
# x와 y값들을 이용하여 격자를 생성
# xx,yy 두개의 2차원 배열이 생성(x,y의 모든 조합에 대한 좌표)

print(xx)
print(yy)
print(xx.ravel())
#ravel() : 각축의 값들을 1차원 배열로 펼친다
print(yy.ravel())
```

```python
import matplotlib.pyplot as plt

plt.scatter(xx, yy)
plt.grid(True)
plt.show()
```
