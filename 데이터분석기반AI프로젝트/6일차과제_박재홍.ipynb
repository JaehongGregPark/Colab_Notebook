```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.rc('font', family='NanumBarunGothic')
plt.rcParams['axes.unicode_minus'] =False
```

```python
# 데이터셋 불러오기
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/AIProject2/day06/rent.csv', encoding='utf-8')
# 관련 열 추출
columns_to_extract = ['Posted On', 'BHK', 'Rent', 'Size', 'Floor', 'Area Type', 'Area Locality', 'City',
                       'Furnishing Status', 'Tenant Preferred', 'Bathroom', 'Point of Contact']
df = data[columns_to_extract]


df.info()
```

```python
# 클러스터링에 사용할 관련 특성 선택
#features = df[['BHK', 'Rent', 'Size', 'Bathroom']]
features = df[[ 'Rent', 'Size']]
```

```python

# 특성 표준화
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# 클러스터 수 선택
n_clusters = 4

# K-means 클러스터링 적용
k_cluster = KMeans(n_clusters=n_clusters, random_state=0)
k_cluster.fit(scaled_features)

k_cluster

```

```python
print(k_cluster.labels_)
print(type(k_cluster.labels_))
```

```python
import numpy as np

print(np.shape(k_cluster.labels_))
print(np.unique(k_cluster.labels_))
```

```python
#중심점 가져오기
centers = k_cluster.cluster_centers_

```
