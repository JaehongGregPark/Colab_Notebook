## ë°ì´í„° íŒŒì´í”„ ë¼ì¸ ìƒì„±í•˜ê¸°

ë°ì´í„° íŒŒì´í”„ ë¼ì¸ì€ ë³¸ ê³¼ì •ì˜ í•µì‹¬ì´ ì•„ë‹ˆê¸° ë•Œë¬¸ì— ì´ëŸ° ì˜ˆì œê°€ ìˆë‹¤ëŠ” ì •ë„ë§Œ ë°›ìœ¼ë“¤ì´ì‹œë©´ ë©ë‹ˆë‹¤. ğŸ˜Š

ë°ì´í„°ë¥¼ ë””ë ‰í† ë¦¬ë¡œë¶€í„° ë¶ˆëŸ¬ì˜¬ ë•Œ í•œë²ˆì— ê°€ì ¸ì˜¬ ë°ì´í„°ì˜ ìˆ˜ì¸ batch sizeë¥¼ ì„¤ì •í•˜ê³ , data generatorë¥¼ ìƒì„±í•˜ì—¬ ë°ì´í„°ë¥¼ ëª¨ë¸ì— ë„£ì„ ìˆ˜ ìˆë„ë¡ í•©ë‹ˆë‹¤.

ìš°ì„  ImageDataGenerator ê°ì²´ë¥¼ ìƒì„±í•˜ì—¬ ë°ì´í„° íŒŒì´í”„ ë¼ì¸ì„ ë§Œë“­ë‹ˆë‹¤. train datasetì˜ generatorì—ëŠ” augmentationì´ í¬í•¨ë˜ê³ , test datasetì˜ generatorëŠ” ì›ë³¸ ìƒíƒœë¥¼ ìœ ì§€í•˜ê³  rescalingë§Œ ì ìš©í•©ë‹ˆë‹¤.

```python
### data íŒŒì´í”„ ë¼ì¸ ìƒì„±

# ë°ì´í„°ë¥¼ ë””ë ‰í† ë¦¬ë¡œë¶€í„° ë¶ˆëŸ¬ì˜¬ ë•Œ, í•œë²ˆì— ê°€ì ¸ì˜¬ ë°ì´í„°ì˜ ìˆ˜
batch_size=20

# Training ë°ì´í„°ì˜ augmentation íŒŒì´í”„ ë¼ì¸ ë§Œë“¤ê¸°
augmentation_train_datagen = tf.keras.preprocessing.image.ImageDataGenerator( rescale=1./255,  # ëª¨ë“  ë°ì´í„°ì˜ ê°’ì„ 1/255ë¡œ ìŠ¤ì¼€ì¼ ì¡°ì •
                                    rotation_range=40,  # 0~40ë„ ì‚¬ì´ë¡œ ì´ë¯¸ì§€ íšŒì „
                                    width_shift_range=0.2,  # ì „ì²´ ê°€ë¡œ ê¸¸ì´ë¥¼ ê¸°ì¤€ìœ¼ë¡œ 0.2 ë¹„ìœ¨ê¹Œì§€ ê°€ë¡œë¡œ ì´ë™
                                    height_shift_range=0.2,   # ì „ì²´ ì„¸ë¡œ ê¸¸ì´ë¥¼ ê¸°ì¤€ìœ¼ë¡œ 0.2 ë¹„ìœ¨ê¹Œì§€ ê°€ë¡œë¡œ ì´ë™
                                    shear_range=0.2,  # 0.2 ë¼ë””ì•ˆ ì •ë„ê¹Œì§€ ì´ë¯¸ì§€ë¥¼ ê¸°ìš¸ì´ê¸°
                                    zoom_range=0.2, # í™•ëŒ€ì™€ ì¶•ì†Œì˜ ë²”ìœ„ [1-0.2 ~ 1+0.2 ]
                                    horizontal_flip=True,)  # ìˆ˜í‰ ê¸°ì¤€ í”Œë¦½ì„ í•  ì§€, í•˜ì§€ ì•Šì„ ì§€ë¥¼ ê²°ì •

# Test ë°ì´í„°ì˜ augmentation íŒŒì´í”„ ë¼ì¸ ë§Œë“¤ê¸°
test_datagen = tf.keras.preprocessing.image.ImageDataGenerator(rescale=1./255)
```

Augmentation íŒŒì´í”„ë¼ì¸ì„ ê¸°ì¤€ìœ¼ë¡œ ë””ë ‰í† ë¦¬ë¡œë¶€í„° ë°ì´í„°ë¥¼ ë¶ˆëŸ¬ì˜¤ëŠ” ëª¨ë“ˆì„ ë§Œë“­ë‹ˆë‹¤. train/test datasetì˜ ê²½ë¡œì™€ ì´ë¯¸ì§€ í¬ê¸° ë“±ì„ ì§€ì •í•˜ì—¬ train_generatorì™€ test_generator ìƒì„±í•˜ë©´ ë©ë‹ˆë‹¤.

```python
# Augmentation íŒŒì´í”„ë¼ì¸ì„ ê¸°ì¤€ìœ¼ë¡œ ë””ë ‰í† ë¦¬ë¡œë¶€í„° ë°ì´í„°ë¥¼ ë¶ˆëŸ¬ ì˜¤ëŠ” ëª¨ë“ˆ ë§Œë“¤ê¸°
train_generator = augmentation_train_datagen.flow_from_directory(
        directory=train_dir, #  ì–´ëŠ ë””ë ‰í„°ë¦¬ì—ì„œ ì´ë¯¸ì§€ ë°ì´í„°ë¥¼ ê°€ì ¸ì˜¬ ê²ƒì¸ê°€?
        target_size=(150, 150), # ëª¨ë“  ì´ë¯¸ì§€ë¥¼ 150 Ã— 150 í¬ê¸°ë¡œ ë°”ê¿‰ë‹ˆë‹¤
        batch_size=batch_size, # ë””ë ‰í† ë¦¬ì—ì„œ batch sizeë§Œí¼ì˜ ì´ë¯¸ì§€ë¥¼ ê°€ì ¸ì˜µë‹ˆë‹¤.
        interpolation='bilinear',  # resizeë¥¼ í•  ë•Œ, interpolatrion ê¸°ë²•ì„ ê²°ì •í•©ë‹ˆë‹¤.
        color_mode ='rgb',
        shuffle='True', # ì´ë¯¸ì§€ë¥¼ ì…”í”Œë§í•  ì§€ í•˜ì§€ ì•Šì„ ì§€ë¥¼ ê²°ì •.
        class_mode='categorical') # multiclassì˜ ê²½ìš°ì´ë¯€ë¡œ class modeëŠ” categorical

print(train_generator.class_indices)
```

```python
# Test ë°ì´í„° ë˜í•œ ë””ë ‰í† ë¦¬ë¡œë¶€í„° ì´ë¯¸ì§€ë¥¼ ë¶ˆëŸ¬ì˜¤ëŠ” íŒŒì´í”„ë¼ì¸ì„ ì™„ì„±í•´ì•¼ í•œë‹¤.
test_generator = test_datagen.flow_from_directory(
        directory=test_dir,
        target_size=(150, 150),
        batch_size=batch_size,
        shuffle='True',
        interpolation='bilinear',  ## resizeì‹œ interpolatrion ê¸°ë²•
        color_mode='rgb',
        class_mode='categorical') #categorical
```

Train dataì˜ íŒŒì´í”„ ë¼ì¸ì´ batch sizeë§Œí¼ì˜ ë°ì´í„°ë¥¼ ì˜ ë¶ˆëŸ¬ì˜¤ëŠ” ì§€ í™•ì¸í•´ ë´…ì‹œë‹¤.

```python
for data_batch, labels_batch in train_generator:
    print('ë°°ì¹˜ ë°ì´í„° í¬ê¸°:', data_batch.shape)
    print('ë°°ì¹˜ ë ˆì´ë¸” í¬ê¸°:', labels_batch.shape)
    break
```

ë°”íƒ•ì´ ë˜ëŠ” Pretrained Model(ResNet50)ì„ ë¶ˆëŸ¬ì˜¤ê³  ëª¨ë¸ì˜ êµ¬ì¡°ë¥¼ ì‚´í´ë´…ì‹œë‹¤.

```python
## back bone
conv_base=tf.keras.applications.ResNet50(weights='imagenet',include_top=False)
conv_base.summary()
```

input layerì™€ ResNet50 backbone, fully-connected layerë¥¼ ì—°ê²°í•˜ì—¬ transfer learning ëª¨ë¸ì„ ë§Œë“­ë‹ˆë‹¤.

```python
# ìµœì¢… ëª¨ë¸ êµ¬ì„±í•˜ê¸°
input_layer = tf.keras.layers.Input(shape=(150,150,3))
x = conv_base(input_layer) # ìœ„ì—ì„œ ë¶ˆëŸ¬ì˜¨ pretrained modelì„ í™œìš©í•˜ê¸°
# ë¶ˆëŸ¬ì˜¨ conv_base ëª¨ë¸ì˜ ìµœì¢… ê²°ê³¼ë¬¼ì€ Conv2D ì—°ì‚°ì˜ feature mapê³¼ ë™ì¼
# ë”°ë¼ì„œ ìµœì¢…ì ì¸ Multiclass classficationì„ í•˜ê¸° ìœ„í•´ì„œëŠ” Flattenì„ í•´ì•¼ í•œë‹¤.
x = tf.keras.layers.Flatten()(x)
x = tf.keras.layers.Dense(512, activation='relu')(x)
out_layer = tf.keras.layers.Dense(4, activation='softmax')(x)
```
